```python
import pandas as pd
```

```python
df_a=pd.read_csv('5_a.csv')
df_a.head()
```

```python
df_a['y_pred'] = df_a['proba'].apply(lambda x: 1  if x>0.5 else 0)

#f"{x} is even" if x %2==0 else f"{x} is odd"
#train_all['comment_text'] = train_all['comment_text'].apply(lambda x: text_cleaning(x))
```

```python
#To Calculate Confusion Matrix, we need to calculkate TP, FP, FN, TN
def calc_confusion_matrix(y_actual, y_pred):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    #confusion_matrix = [[TP, FP], [FN, TN]]
    for i in range(len(df_a)):
        if y_actual[i]==1 and y_pred[i] == 1:
            TP +=1
        if y_actual[i] ==0 and y_pred[i] ==0:
            TN +=1
        if y_actual[i] == 0 and y_pred[i]==1:
            FP +=1
        if y_actual[i]==1 and y_pred[i]== 0:
            FN +=1
        
    return {'TP': TP , 'FP': FP, 'FN': FN, 'TN': TN}

```

```python
confusion_matrix = calc_confusion_matrix(df_a.y, df_a.y_pred)
```

```python
confusion_matrix
confusion_matrix_matrix = [[confusion_matrix['TP'], confusion_matrix['FP']], [confusion_matrix['FN'], confusion_matrix['TN']]]
```

```python
confusion_matrix
```

```python
confusion_matrix_matrix
```

<h4>F1 Score</h4>

```python
precision = confusion_matrix['TP'] / (confusion_matrix['TP'] + confusion_matrix['FP'])
recall = confusion_matrix['TP'] / (confusion_matrix['TP'] + confusion_matrix['FN'])
```

```python
f1_score = 2*precision* recall / (precision + recall)
```

```python
print('The F1 score is: ', f1_score)
```

<h4>AUC Score</h4>

```python
def perf_metrics(df, y, proba):
    # get number of positive and negative examples in the dataset
    P = sum(y)
    N = len(y) - P
    TP = 0
    FP = 0
    TPR = []
    FPR = []
    sorted_df = df.sort_values(by=['proba'], ascending=False)

    thresholds = sorted_df['proba'].unique()

    #thresholds = [0,.05,.1,.15,.2,.25,.3,.35,.4,.45,.5,.55,.6,.65,.7,.75,.8,.85,.9,.95,1]
    for thresh in thresholds:
        FP=0
        TP=0
        thresh = round(thresh,2) #Limiting floats to two decimal points, or threshold 0.6 will be 0.6000000000000001 which gives FP=0
        for i in range(len(proba)): 
            if(proba[i] >= thresh):
                if(y[i] == 1):
                    TP += 1
                else:
                    FP += 1

        #We find the True positive rate and False positive rate based on the threshold
                
        TPR.append(TP/P)
        FPR.append(FP/N)

    return {'FPR' : FPR, 'TPR': TPR, 'AUC' :  np.trapz(TPR,FPR)}
```

```python
FPR_TPR_AUC = perf_metrics(df_a, df_a.y, df_a.proba)
```

```python
print('The AUC score is: ', FPR_TPR_AUC['AUC'])
```

<h4>Accuracy</h4>

```python
# Accuracy
Accuracy = (confusion_matrix['TP']+confusion_matrix['TN'])/df_a.shape[0]
print('The accuracy is: ',Accuracy)
```

### C. Compute the best threshold (similarly to ROC curve computation) of probability which gives lowest values of metric <b>A</b> for the given data 
<br>

you will be predicting label of a data points like this: $y^{pred}= \text{[0 if y_score < threshold  else 1]}$

$ A = 500 \times \text{number of false negative} + 100 \times \text{numebr of false positive}$

<pre>
   <b>Note 1:</b> in this data you can see number of negative points > number of positive points
   <b>Note 2:</b> use pandas or numpy to read the data from <b>5_c.csv</b>
</pre>

```python
df_c=pd.read_csv('5_c.csv')
df_c.head()
```

```python
sorted_df = df_c.sort_values(by=['prob'], ascending=False)
thresholds_df_c = sorted_df['prob'].unique()
```

```python
 # write your code for task C
def min_A(y, proba, threshold):
    # get number of positive and negative examples in the dataset

    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(proba)): 
        if(proba[i] >= threshold):
            if(y[i] == 1):
                TP += 1
            else:
                FP += 1
        if proba[i] < threshold:
            if y[i] == 0:
                TN += 1
            else:
                FN += 1


        #We find the True positive rate and False positive rate based on the threshold
                
    A = (500 * FN) + (100 * FP)    
        

    return A
```

```python
def get_minimized_a(y_actual, y_probabilities, total_threshold_arr ):
  min_a = float('inf')
  min_t = 0

  for threshold in total_threshold_arr:
    a = min_A(y_actual, y_probabilities, threshold)
    if a <= min_a:
        min_a = min(a, min_a)
        min_t = threshold

  return min_a, min_t
```

```python
print(get_minimized_a(df_c.y, df_c.prob, thresholds_df_c))
# (141000, 0.2300390278970873)
```

<h4>Mean Square Error</h4>

```python
def MSE(df, y, y_pred):
    MSE = 1
    for i in range(len(df)):
        difference = y[i] - y_pred[i]
        square = difference**2
        MSE += square
    return MSE/len(df)
```

```python
MSE = MSE(df_d, df_d.y, df_d.pred)
```

```python
MSE
```

<h4>Mean Absolute Percantage Error</h4>

```python
def MAPE(df, y, pred):
    # get number of positive and negative examples in the dataset
    sum = 1
    y_mean = np.mean(y)
    for i in range(len(df)): 

        error = (y[i] - pred[i])/ y_mean
        abs_error = np.abs(error)
        sum += abs_error
    return (sum*100)/len(df)
```

```python
MAPE = MAPE(df_d, df_d.y, df_d.pred)
```

```python
print('The MAPE value is', MAPE)
```

<h4>R^2 Error</h4>

```python
def r2_square(df, y, pred):
    # get number of positive and negative examples in the dataset
    SSR = 1
    SST = 1
    y_mean = np.mean(y)
    for i in range(len(df)): 
        
        error = y[i] - pred[i]
        error_square = error**2
        SSR += error_square

        mean_error = y[i] - y_mean
        mean_error_square = mean_error**2
        SST += mean_error_square

    return 1 - (SSR/SST)
```

```python
r2_square = r2_square(df_d, df_d.y, df_d.pred)
```

```python
print('The Co-efficient of determination value is', r2_square)
```
